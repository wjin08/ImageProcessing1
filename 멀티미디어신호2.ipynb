```python
import cv2
from google.colab.patches import cv2_imshow
import sys
import numpy as np
src=cv2.imread('/content/drive/MyDrive/Colab Notebooks/ss.jpg')

if src is None:
  print('Img load fail')
  sys.exit()

dst=cv2.bilateralFilter(src, -1,10,5)

cv2_imshow(src)
cv2_imshow(dst)
cv2.waitKey()

cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
hist=cv2.calcHist([dst],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()

img=cv2.imread('/content/drive/MyDrive/Colab Notebooks/ss.jpg')
channels=cv2.split(img)
colors= ['b','g','r']
for ch, color in zip(channels, colors):
  hist=cv2.calcHist([ch],[0],None,[256],[0,256])
  plt.plot(hist, color=color)
plt.show()
```

```python
import cv2
from google.colab.patches import cv2_imshow
import sys
import numpy as np
import argparse
import os
import matplotlib.pyplot as plt

path = '/content/drive/MyDrive/Colab Notebooks/ss.jpg'

src=cv2.imread(path,cv2.IMREAD_GRAYSCALE)
dst=cv2.equalizeHist(src)
srcHist=cv2.calcHist(images=[src],
                     channels=[0],
                     mask=None,
                     histSize=[256],
                     ranges=[0,256])
dstHist=cv2.calcHist(images=[dst],
                     channels=[0],
                     mask=None,
                     histSize=[256],
                     ranges=[0,256])
cv2_imshow(src)
cv2_imshow(dst)

plt.plot(srcHist, color ='b', label='src hist')
plt.plot(dstHist, color='r', label='dst hist')
plt.legend(loc='best')
plt.show()
cv2.waitKey()
cv2.destroyAllWindows()

```

```python

import numpy as np
import cv2
import matplotlib.pyplot as plt

img = np.array([[3, 2, 2, 2, 2, 3, 3, 4],
                [3, 2, 2, 2, 3, 4, 3, 3],
                [4, 3, 3, 4, 4, 4, 3, 3],
                [5, 4, 4, 4, 5, 4, 3, 3],
                [5, 4, 3, 4, 5, 4, 3, 2],
                [6, 5, 4, 4, 5, 4, 3, 2],
                [6, 6, 5, 5, 4, 3, 2, 2],
                [6, 5, 4, 5, 4, 3, 2, 2]])

hist, bins = np.histogram(img.flatten(), 256, [0, 256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()

cdf_normalized = np.uint8(cdf_normalized)
img_equalized = cdf_normalized[img]


plt.hist(img.flatten(), 256, [0, 256], color = 'r')
plt.xlim([0, 256])
plt.show()


```
